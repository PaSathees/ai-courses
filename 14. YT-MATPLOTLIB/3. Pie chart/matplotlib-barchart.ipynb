## Matplotlib YouTube Course Notebook - Pie Chart

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
slices = [120, 80, 30, 20]
labels =['sixty', 'forty', 'yellow', 'green']
colors = ['blue', 'red', 'yellow', 'green']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f

slices = [120, 80, 30, 20]
labels = ['sixty', 'forty', 'yellow', 'green']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors,
        wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
plt.pie(slices, labels=labels, colors=colors,
        wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] # radius


plt.pie(slices, labels=labels, colors=colors, explode=explode,
        wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.pie(slices, labels=labels, colors=colors, explode=explode, shadow=True,
        wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.pie(slices, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90,
        wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.pie(slices, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90, autopct="%1.1f%%",
        wedgeprops={'edgecolor': 'black'})
plt.title("My awesome Pie Chart")
plt.tight_layout()
plt.show()
```
